```python
import requests
import csv
import json

# Load the index settings and mappings from forum_index.json
with open('forum_index.json', 'r', encoding='utf-8') as file:
    index_settings = json.load(file)

# Define the Elasticsearch URL and headers
es_url = 'http://localhost:9200/forum'
headers = {'Content-Type': 'application/json'}

# Create the index with defined mappings and settings
response = requests.put(es_url, headers=headers, data=json.dumps(index_settings))
print("Create Index Response:", response.json())

# Load data from output.csv and print for verification
data = []
with open('./FileStore/tables/output.csv', 'r', encoding='utf-8') as file:
    reader = csv.reader(file)
    for row in reader:
        userid, name, message = row
        print(f"UserId: {userid}, Name: {name}, Message: {message}")
        data.append({
            "UserId": int(userid),
            "Name": name,
            "Message": message
        })

# Define Elasticsearch document URL
es_doc_url = 'http://localhost:9200/forum/_doc'

# Index each record in Elasticsearch and print response
for record in data:
    response = requests.post(es_doc_url, headers=headers, data=json.dumps(record))
    print("Index Document Response:", response.json())

# Verify indexed data
search_url = 'http://localhost:9200/forum/_search'
search_query = {
    "query": {
        "match_all": {}
    },
    "size": 5  # Limit the response to 5 documents
}
response = requests.get(search_url, headers={'Content-Type': 'application/json'}, data=json.dumps(search_query))
print("Verify Indexed Data Response:", response.json())
```
